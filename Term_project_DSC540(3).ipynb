```python
print(df.columns)
df=df.rename(columns={'datetimeStr':'date'})
df.columns
```

```python
#selecting columns containing date, temperature, minimum temperature, and maximum temperature
df1=df[["temp", "maxt", "date", "mint", "wspd"]]
df1
```

```python
df1.rename(columns={"maxt":"max_temp", "mint":"min_temp"}, inplace=True)
df1
```

```python
#adding new column year
df1["year"]=[int(i[:4]) for i in df1["date"]]
df1
```

```python
#adding new columns: day and month
month=[int(i[5:7]) for i in df1["date"]]
df1["month"]=month
df1["day"]=[int(i[8:10]) for i in df1["date"]]
df1
```

```python
#filtering only yyyy-mo-dd portion of date
import re
pattern=re.compile(r"\d{4}[-]\d{2}[-]\d{2}")
date_yyyy_mm_dd=[pattern.search(i).group() for i in df1["date"]]
df1["date"]=date_yyyy_mm_dd
df1
```

```python
#the month_conversion(month) function converts month number to month name
def month_conversion(month):
    if month==1:
        return "January"
    elif month==2:
        return "February"
    elif month==3:
        return "March"
    elif month==4:
        return "April"
    elif month==5:
        return "May"
    elif month==6:
        return "June"
    elif month==7:
        return "July"
    elif month ==8:
        return "August"
    elif month==9:
        return "September"
    elif month==10:
        return "October"
    elif month==11:
        return "November"
    elif month==12:
        return "December"
            
#changing the month number to the month name 
df1["month"]=df1["month"].apply(month_conversion)
df1.head()
```

```python
#combaining two variables "month" and "day" into one variable month_day
df1["month_day"] = df1["month"]+" "+df1["day"].map(str)
print(df1.head())
```

```python
"""changing the temperature to Kelvins and windspeed to m/s
T(kelvins)=(T(F)  32)  5/9+273
V(km/h)=3.6*V(m/s)"""
df1["temp"]=round(5*(df1["temp"]-32)/9+273, 2)
df1["max_temp"]=round(5*(df1["max_temp"]-32)/9+273, 2)
df1["min_temp"]=round(5*(df1["min_temp"]-32)/9+273, 2)
df1["wind_speed"]=[round(i/3, 2) for i in df1["wspd"]]
df1
```

```python
#February 29 should be removed from the df4 data frame 
print(df1.set_index('month_day').index.get_loc("February 29"))
```

```python
df1.drop(df1.index[883], inplace=True)
df1.drop(df1.index[0], inplace=True)
df1.reset_index(drop=True)
```

```python
#subsetting the data based on years and resetting the index
df1_2017=df1[(df1["year"]==2017)].reset_index(drop=True)
df1_2018=df1[(df1["year"]==2018)].reset_index(drop=True)
df1_2019=df1[(df1["year"]==2019)].reset_index(drop=True)
df1_2020=df1[(df1["year"]==2020)].reset_index(drop=True)
print(df1_2017.shape)
print(df1_2018.shape)
print(df1_2019.shape)
print(df1_2020.shape)
print(df1_2020.head())
```

```python
#checking missing values
for col in df1.columns:
    missing_values=df1[col].isnull().sum()
    if missing_values >0:
        print("{} has {} missing value(s)!".format(col, missing_values))
    else:
        print("{} has no missing values".format(col))
```

```python
#calculating index of "September 29", and "December 31"
print(df1_2018.set_index('month_day').index.get_loc("September 29"))

print(df1_2018.set_index('month_day').index.get_loc("December 31"))
```

```python
df1_2017_2018=df1_2017.append(df1_2018.loc[:271]).reset_index(drop=True)
df1_2017_2018=df1_2017_2018[["temp", "max_temp", "min_temp", "wind_speed", "month_day"]]
df1_2017_2018.columns=["temp_2017_2018","max_temp_2017_2018", "min_temp_2017_2018", "wind_speed_2017_2018", "month_day"]
df1_2017_2018
```

```python
df1_2018_2019=df1_2018.loc[272:].append(df1_2019.loc[:271]).reset_index(drop=True)
df1_2018_2019=df1_2018_2019[["temp", "max_temp", "min_temp", "wind_speed", "month_day"]]
df1_2018_2019.columns=["temp_2018_2019","max_temp_2018_2019", "min_temp_2018_2019", "wind_speed_2018_2019", "month_day"]
df1_2018_2019
```

```python
df1_2019_2020=df1_2019.loc[272:].append(df1_2020).reset_index(drop=True)
df1_2019_2020=df1_2019_2020[["temp", "max_temp", "min_temp", "wind_speed", "month_day"]]
df1_2019_2020.columns=["temp_2019_2020","max_temp_2019_2020", "min_temp_2019_2020", "wind_speed_2019_2020", "month_day"]
df1_2019_2020
```

```python
merged_2017_2020=pd.merge(df1_2017_2018, df1_2018_2019, on="month_day", how="outer" )
merged_2017_2020=pd.merge(merged_2017_2020, df1_2019_2020, on="month_day", how="outer" )
merged_2017_2020
```
