```python
"""They provided data for each hour per a day, but we need only a row per a day. 
Changing the "dt_iso" variable to the "date" variable"""
df2=df2.rename(columns = {'dt_iso':'date'})
df2.columns
```

```python
import re
print(df2["date"][:24])
```

```python
"""Because https://openweathermap.org/ provided historical weather data for each hour per a day, but we need 
only one record per a day all data except 12:00:00 will be filtered out"""
pattern=re.compile(r"(?<=\s)(12)(?=:)")
df3=df2[df2["date"].str.contains(pattern)]
df3.head()
```

```python
"""The date column contains data in format yyyy-mm-dd time ...We need to filter out everything except only 
date:yyyy-mm-dd. It can be done by two ways. In the first way we are extracting the first 10 elements of each string in
the date column. In the second approach we use regular expressions to extract required pattern"""
# date_yyyy_mm_dd=[i[0:10] for i in df3["date"]]
# df3["date"]=date_yyyy_mm_dd
# df3.head()

pattern=re.compile(r"\d{4}[-]\d{2}[-]\d{2}")
date_yyyy_mm_dd=[pattern.search(i).group() for i in df3["date"]]
df3["date"]=date_yyyy_mm_dd
```

```python
#creating the year variable that contains only year part of date
years=[int(i[:4]) for i in df3["date"]]
df3["years"]=years
```

```python
#creating the month variable that contains only the month part of the date
month=[int(i[5:7]) for i in df3["date"]]
# print(month)
df3["month"]=month
#the month_conversion(month) function converts month number to month name
def month_conversion(month):
    if month==1:
        return "January"
    elif month==2:
        return "February"
    elif month==3:
        return "March"
    elif month==4:
        return "April"
    elif month==5:
        return "May"
    elif month==6:
        return "June"
    elif month==7:
        return "July"
    elif month ==8:
        return "August"
    elif month==9:
        return "September"
    elif month==10:
        return "October"
    elif month==11:
        return "November"
    elif month==12:
        return "December"
            
#changing the month number to the month name 
df3["month"]=df3["month"].apply(month_conversion)
print(df3.head())
```

```python
#creating the day variable that contains only the day portion of the date
day=[int(i[8:10]) for i in df3["date"]]
# print(day)
df3["day"]=day
print(df3.head())
df3.drop_duplicates(subset=["date"])
df3.shape                  
```

```python
#combaining two columns into one
df3["month_day"] = df3["month"]+" "+df3["day"].map(str)
print(df3.head())
```
