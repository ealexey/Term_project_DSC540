```python
#downloading the weather_data_1hr.csv file
df=pd.read_csv("weather_data_1hr.csv")
df.head(25)
```

```python
print("The df dataset has {} columns and {} rows".format(df.shape[0], df.shape[1]))
df.columns
```

```python
"""Because https://www.worldweatheronline.com/ provided historical weather data for each hour per a day, but we need 
only one record per a day all data except 1200(noon) will be filtered out"""
df1=df[df["time"]==1200].reset_index(drop=True)
df1.head()
```

```python
print("The df1 dataset has {} columns and {} rows".format(df1.shape[0], df1.shape[1]))
```

```python
# We need to filter only dates 09/30/2011-09/29/2014
print(df1.set_index('date').index.get_loc("2011-09-30"))

print(df1.set_index('date').index.get_loc("2014-09-29"))
```

```python
df2=df1.loc[1186:2281].reset_index(drop=True)
print(df2.head())
print(df2.tail())
```

```python
#checking missing values

for col in df2.columns:
    missing_values=df2[col].isnull().sum()
    if missing_values>0:
        print("{} has {} missing value(s)".format(col, missing_values))
    else:
        print("{} has no missing values".format(col))
```

```python
#selecting rows for further analysis
df3=df2[["date", "tempC", "windspeedKmph"]]
df3
```

```python
"""changing the temperature to Kelvins and windspeed to m/s
T(kelvins)=T(celsius)+273
V(km/h)=3.6*V(m/s)"""
df3["temp"]=df3["tempC"]+273
df3["wind_speed"]=[round(i/3, 2) for i in df3["windspeedKmph"]]
df3
```

```python
df3=df3.drop(columns=["tempC", "windspeedKmph"])
df3
```

```python
#February 29 should be removed from the df3 data frame 
print(df3.set_index('date').index.get_loc("2012-02-29"))


```

```python

df3.drop(df3.index[152], inplace=True)
df3.reset_index(drop=True)
```

```python
df3["year"]=[int(i[:4]) for i in df3["date"]]
df3
```

```python
#generation of two columns: "month" and "day"
df3["month"]=[int(i[5:7]) for i in df3["date"]]
df3["day"]=[int(i[8:10]) for i in df3["date"]]
df3
```

```python
#transformation of month number to month name
def month_conversion(month):
    if month==1:
        return "January"
    elif month==2:
        return "February"
    elif month==3:
        return "March"
    elif month==4:
        return "April"
    elif month==5:
        return "May"
    elif month==6:
        return "June"
    elif month==7:
        return "July"
    elif month ==8:
        return "August"
    elif month==9:
        return "September"
    elif month==10:
        return "October"
    elif month==11:
        return "November"
    elif month==12:
        return "December"
```

```python
df3["month"]=df3["month"].apply(month_conversion)
df3.head()
```

```python
#generation of new column month_day
df3["month_day"]=df3["month"]+" "+df3["day"].map(str)
df3
```

```python
#generation of new dataframes containing data for corresponding year
df3_2011=df3[df3["year"]==2011].reset_index(drop=True)
df3_2012=df3[df3["year"]==2012].reset_index(drop=True)
df3_2013=df3[df3["year"]==2013].reset_index(drop=True)
df3_2014=df3[df3["year"]==2014].reset_index(drop=True)
```

```python
# #calculating index of "September 30", and "December 31" for 2012
print(df3_2012.set_index('month_day').index.get_loc("September 29"))

print(df3_2012.set_index('month_day').index.get_loc("December 31"))
```

```python
df3_2011_2012=df3_2011.append(df3_2012.loc[:271]).reset_index(drop=True)
df3_2011_2012=df3_2011_2012[["temp", "wind_speed", "month_day"]]
df3_2011_2012.columns=["temp_2011_2012", "wind_speed_2011_2012", "month_day"]
df3_2011_2012
```

```python
df3_2012_2013=df3_2012.loc[272:].append(df3_2013.loc[:271]).reset_index(drop=True)
df3_2012_2013=df3_2012_2013[["temp", "wind_speed", "month_day"]]
df3_2012_2013.columns=["temp_2012_2013", "wind_speed_2012_2013", "month_day"]
df3_2012_2013
```

```python
df3_2013_2014=df3_2013.loc[272:].append(df3_2014).reset_index(drop=True)
df3_2013_2014=df3_2013_2014[["temp", "wind_speed", "month_day"]]
df3_2013_2014.columns=["temp_2013_2014", "wind_speed_2013_2014", "month_day"]
df3_2013_2014
```

```python
merged_2011_2014=pd.merge(df3_2011_2012, df3_2012_2013, on="month_day", how="outer" )
merged_2011_2014=pd.merge(merged_2011_2014, df3_2013_2014, on="month_day", how="outer" )
merged_2011_2014
```
