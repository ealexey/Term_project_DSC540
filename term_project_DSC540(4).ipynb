```python
%pwd
import pandas as pd
import numpy as np
```

```python
#importing the merged_2011_2014.pkl data frame from milestone 2 
df1=pd.read_pickle("merged_2011_2014.pkl")
df1.head(10)
```

```python
#importing the merged_2014_2017.pkl data frame from milestone 3
df2=pd.read_pickle("merged_2014_2017.pkl")
df2
```

```python
##importing the merged_2017_2020.pkl data frame from milestone 4
df3=pd.read_pickle("merged_2017_2020.pkl")
df3
```

```python
#merging all 3 dataframes from milestone 2, 3, and 4
merged=pd.merge(df1, df2, on="month_day", how="outer" )
merged=pd.merge(merged, df3, on="month_day", how="outer")
merged
```

```python
"""plitting the column month_day into two columns (the first way) the month column will be used for grouping using
the groupby() method"""
import re
pattern=re.compile(r"[a-zA-Z]+") #pattern to exctract month
month=[pattern.search(i).group() for i in merged["month_day"]]
pattern1=re.compile(r"\d+") #pattern to exctract day
day=[pattern1.search(i).group() for i in merged["month_day"]]
merged["month"]=month
merged["day"]=day
merged.drop(columns=["month_day"], inplace=True)
merged
                    
                    
```

```python
#the merged dataset contains some numbers with different amount of decimal numbers, for example 7.999999998 and 3.45 
#in spite of rounding data in all datasets. The merded dataset will be cleaned again
a=(len(merged.columns)-3)
for col in merged.columns[:a]:
    merged[col]=([round(i,2) for i in merged[col]])
merged

```

```python
#creating a new column that correspond to the month number to sort data frame rows from January (1) to December (12)
def month_to_number(month):
    if month=="January":
        return 1
    elif month=="February":
        return 2
    elif month=="March":
        return 3
    elif month=="April":
        return 4
    elif month=="May":
        return 5
    elif month=="June":
        return 6
    elif month=="July":
        return 7
    elif month=="August":
        return 8
    elif month=="September":
        return 9
    elif month=="October":
        return 10
    elif month=="November":
        return 11
    elif month=="December":
        return 12
merged["month_number"]=merged["month"]
merged["month_number"]=merged["month_number"].apply(month_to_number)
merged.drop(columns=["day"], inplace=True)
merged
                                                                          
                                                                          
            
```

```python
#histograms for numerical variables in the merged dataframe
import matplotlib.pyplot as plt
for col in merged.columns[:a]:
    plt.title("Histogram of "+col,fontsize=15)
    plt.hist(merged[col].dropna(), bins=20)
    plt.show()
```

```python
#boxplots for numerical variables in the merged dataframe
for col in merged.columns[:a]:
    merged.boxplot(col)
    plt.title(col)
    plt.show()
    
```

```python
"""generation of the dataframe by_month["temp_2011_2012"]. This dataframe contains month's names as index
and has two columns. The first column indicates the mean temperature for corresponding month and the second
column indicates the standard deviation for the corresponding month"""
merged.sort_values(["month_number"], inplace=True)
by_month=merged.groupby("month", sort=False).agg([np.mean, np.std])
by_month["temp_2011_2012"]
```

```python
"""bar plot with errorbars. Each bar with errorbar corresponds to the mean temperature for the corresponding month
The period encompasses dates from September 30, """ 
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_axes([0,0,2,1])
month = by_month["temp_2011_2012"]
ax.set_title("The mean monthly temperature from 09/30/2011 to 09/29/2012")
ax.bar(x=month.index, height=month[month.columns[0]]-273, yerr=month[month.columns[1]], capsize=6)
plt.show()

```

```python
"""grouped bar chart with errorbars. This bar chart shows mean temperature for each months for two periods
1. 09/30/2011 to 09/29/2012
2. 09/30/2018 to 09/29/2019
"""
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_axes([0,0,2,1])
month = by_month["temp_2011_2012"]
month1=by_month["temp_2018_2019"]
width=0.25
labels=month.index
x=np.arange(len(labels))
labels=month.index
ax.set_ylabel('Temperature C')
ax.set_title('Temperature of 2011-2012 and 2018-2019')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.bar(x-width/2, month[month.columns[0]]-273,  width, yerr=month[month.columns[1]], capsize=6, label="2011-2012")
ax.bar(x+width/2, month1[month1.columns[0]]-273, width, yerr=month1[month1.columns[1]],  capsize=6, label="2018-2019")
ax.legend()
plt.show()
```

```python
"""Scatter plot wind speed vs temperature for the period from 09/30/2019 to 09/29/2020.
This graph shows that there is no correlation between wind speed and temperature"""
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure()
ax = fig.add_axes([0,0,2,1])
ax.scatter(merged["temp_2019_2020"], merged["wind_speed_2019_2020"])

ax.set_xlabel('Temperature')
ax.set_title('Temperature vs wind speed 2019-2020')
ax.set_ylabel("Wind speed")
```

```python
#selecting data for January only from the merged data frame
#we are going to compare mean January temperature from 2012 to 2020
january_temp=merged[merged["month"]=="January"].reset_index(drop=True)
aggr=january_temp.agg([np.mean, np.std]).drop(columns=["month_number"])
aggr.columns
```

```python
#selecting only columns containig temperature data at noon, not max_temp, min_temp, wind_speed
for col in aggr.columns:
    if not str(col).startswith("temp_2"):
        aggr=aggr.drop(columns=[col])
aggr.index[0]
aggr.index[1]
mean=aggr.loc[aggr.index[0]]
std=aggr.loc[aggr.index[1]]
aggr
    
```

```python
#creating a bar plot showing the mean January temperature from 2012 to 2020
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_axes([0,0,2,1])
ax.set_title("The mean January temperature for period from 2012 to 2020")
list_of_years=np.arange(2012, 2021)
ax.bar(x=list_of_years, height=mean-273, yerr=std, capsize=6)
plt.show()
```
